```python
import matplotlib.pyplot as plt
import numpy as np
import glob 
```

```python
block = "SGRGGLGGQGAGFGAILSSGGAGQGGYGGLGSQGT"

polarity = {
    'A': 'np',
    'C': 'np',
    'D': 'np',
    'E': 'np',
    'F': 'np',
    'G': 'np',
    'H': 'np',
    'I': 'np',
    'K': 'np',
    'L': 'np',
    'M': 'np',
    'N': 'p',
    'P': 'np',
    'Q': 'p',
    'R': 'np',
    'S': 'p',
    'T': 'p',
    'V': 'np',
    'W': 'np',
    'Y': 'p'
}

charge = {
    'A': 0,
    'C': 0,
    'D': -1,
    'E': -1,
    'F': 0,
    'G': 0,
    'H': 1,
    'I': 0,
    'K': 1,
    'L': 0,
    'M': 0,
    'N': 0,
    'P': 0,
    'Q': 0,
    'R': 1,
    'S': 0,
    'T': 0,
    'V': 0,
    'W': 0,
    'Y': 0
}

hydrophobicity = {
    'A': 1.8,
    'C': 2.5,
    'D': -3.5,
    'E': -3.5,
    'F': 2.8,
    'G': -0.4,
    'H': -3.2,
    'I': 4.5,
    'K': -3.9,
    'L': 3.8,
    'M': 1.9,
    'N': -3.5,
    'P': -1.6,
    'Q': -3.5,
    'R': -4.5,
    'S': -0.8,
    'T': -0.7,
    'V': 4.2,
    'W': -0.9,
    'Y': -1.3
}
```

```python
charge_distribution = np.array([charge[s] for s in block]).reshape(1,-1)
polarity_distribution = np.array([0 if polarity[s] == 'p' else 1 for s in block]).reshape(1,-1)
hydrophobicity_distribution = np.array([hydrophobicity[s] for s in block]).reshape(1,-1)

fig, ax = plt.subplots(figsize=(20, 10))
im1 = ax.imshow(charge_distribution, cmap='jet')
ax.set_title('Charge Distribution', fontsize=20)
x_ticks_labels = [s for s in block]
ax.set_xticks(list(range(len(block))))
a = ax.set_xticklabels(x_ticks_labels, fontsize=18)
fig.colorbar(im1, orientation='vertical', shrink=0.5)
plt.show()

fig, ax = plt.subplots(figsize=(20, 10))
im2 = ax.imshow(polarity_distribution, cmap='jet')
ax.set_title('Polarity Distribution', fontsize=20)
x_ticks_labels = [s for s in block]
ax.set_xticks(list(range(len(block))))
a = ax.set_xticklabels(x_ticks_labels, fontsize=18)
fig.colorbar(im2, orientation='vertical', shrink=0.5)
plt.show()

fig, ax = plt.subplots(figsize=(20, 10))
im3 = ax.imshow(hydrophobicity_distribution, cmap='jet')
ax.set_title('Hydrophobicity Distribution', fontsize=20)
x_ticks_labels = [s for s in block]
ax.set_xticks(list(range(len(block))))
a = ax.set_xticklabels(x_ticks_labels, fontsize=18)
fig.colorbar(im3, orientation='vertical', shrink=0.5)
plt.show()
```
