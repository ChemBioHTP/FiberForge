```python
import enzy_htp
from enzy_htp.structure import PDBParser
from enzy_htp.mutation import assign_mutant, mutate_stru
from enzy_htp.mutation.api import sync_mutation_over_chains
import random
import numpy as np
from copy import deepcopy
import matplotlib.pyplot as plt

amino_acid = {'CYS': 'C', 'ASP': 'D', 'SER': 'S', 'GLN': 'Q', 'LYS': 'K',
     'ILE': 'I', 'PRO': 'P', 'THR': 'T', 'PHE': 'F', 'ASN': 'N',
     'GLY': 'G', 'HIS': 'H', 'LEU': 'L', 'ARG': 'R', 'TRP': 'W',
     'ALA': 'A', 'VAL':'V', 'GLU': 'E', 'TYR': 'Y', 'MET': 'M'}
```

```python
n_mutants = 1
pdb_file = "6GK3.pdb"
parser = PDBParser()

for m in range(n_mutants):
    structure = parser.get_structure(pdb_file)
    mutations = assign_mutant(
        structure,
        pattern="r:1[chain A:all]*1R",
        chain_sync_list=[tuple(structure.chain_names)],
    )

    mutated_structure = mutate_stru(structure, mutations[0])
    # Save mutated structure
    parser.save_structure(
        stru=mutated_structure, 
        outfile=f"{pdb_file.split('.')[0]}_mutant_{m}_v2.pdb"
    )
    print(f"Mutated structure saved as {pdb_file.split('.')[0]}_mutant_{m}.pdb")
```

```python
from Bio.PDB import PDBParser, PPBuilder, NeighborSearch
from Bio.PDB.Polypeptide import PPBuilder

# Define a function to calculate distance between two atoms
def distance(atom1, atom2):
    return np.linalg.norm(atom1 - atom2)

# Define a function to find hydrogen bonds
def find_hydrogen_bonds(structure):
    ppb = PPBuilder()
    chains = ppb.build_peptides(structure)
    # ns = NeighborSearch(list(structure.get_atoms()))

    hydrogen_bonds = []

    for chain1 in chains:
        for chain2 in chains:
            if chain1 != chain2:
                for residue1 in chain1:
                    for residue2 in chain2:
                        for atom1 in residue1:
                            for atom2 in residue2:
                                # Check if atom1 is donor and atom2 is acceptor or vice versa
                                if (atom1.element == 'N' or atom1.element == 'O') and \
                                   (atom2.element == 'N' or atom2.element == 'O'):
                                    distance_cutoff = 3.5  # You can adjust this value as needed
                                    if distance(np.array(atom1.coord), np.array(atom2.coord)) < distance_cutoff:
                                        hydrogen_bonds.append((atom1, atom2))

    return hydrogen_bonds

# Parse the PDB file
parser = PDBParser()
structure = parser.get_structure("structure", "6GK3.pdb")

# Find hydrogen bonds
hydrogen_bonds = find_hydrogen_bonds(structure)

# Print the hydrogen bonds
for bond in hydrogen_bonds:
    res1 = bond[0]
    res2 = bond[1]
    print(f"Hydrogen bond between {res1.get_full_id()[2]}:{res1.get_parent().id[1]} and {res2.get_full_id()[2]}:{res2.get_parent().id[1]}")
```

```python
import pymol
from pymol import cmd

# Initialize PyMOL
pymol.finish_launching(['pymol', '-qc'])

# Load the PDB file
cmd.load("6GK3.pdb", "structure")

# Define a function to find hydrogen bonds between chains
def find_hydrogen_bonds_between_chains():
    # Find potential hydrogen bonds between chains A and B with a distance cutoff of 3.5 
    pairs = cmd.find_pairs("(chain A)", "(chain B)", cutoff=3.5)
    
    # Iterate over the pairs and filter based on distance
    for pair in pairs:
        pair = pair[0]
        atom1 = f"resi {pair[0].resi} and chain {pair[0].chain} and name {pair[0].name}"
        atom2 = f"resi {pair[1].resi} and chain {pair[1].chain} and name {pair[1].name}"
        dist = cmd.distance(atom1, atom2)
        if dist < 3.5:  # Adjust the distance cutoff as needed
            print(f"Hydrogen bond between {pair[0].resn}{pair[0].resi}.{pair[0].chain} "
                  f"and {pair[1].resn}{pair[1].resi}.{pair[1].chain}")

# Execute the function to find hydrogen bonds between chains
find_hydrogen_bonds_between_chains()

# Delete the loaded structure
cmd.delete("structure")
```

```python
from Bio.PDB import PDBParser
import numpy as np

def hydrogen_bond_plot(pdb_file, chain_id1, chain_id2):
    # Parse the PDB file
    parser = PDBParser()
    structure = parser.get_structure('pdb', pdb_file)

    # Extract the chains
    chain1 = None
    chain2 = None
    for c in structure.get_chains():
        if c.id == chain_id1:
            chain1 = c
        if c.id == chain_id2:
            chain2 = c

    if chain1 is None or chain2 is None:
        raise ValueError("One or both chains not found in the PDB file.")

    # Initialize hydrogen bond plot
    num_residues1 = len(chain1)
    num_residues2 = len(chain2)
    hb_plot = np.zeros((num_residues1, num_residues2), dtype=int)

    # Calculate hydrogen bonds
    for i, residue1 in enumerate(chain1):
        for j, residue2 in enumerate(chain2):
            for atom1 in residue1:
                for atom2 in residue2:
                    # print(atom1.element, atom2.element)
                    # print(atom1, atom2)
                    # print(atom1 - atom2)
                    if (atom1.element == 'O' and atom2.element == 'N') or \
                        (atom1.element == 'N' and atom2.element == 'O'):
                        distance = abs(atom1 - atom2)
                        if distance <= 3.5:  # Threshold for hydrogen bond distance
                            hb_plot[i, j] = 1
                            break
    return hb_plot

def highlight_cell(x,y, ax=None, **kwargs):
    rect = plt.Rectangle((x-.5, y-.5), 1,1, fill=False, **kwargs)
    ax = ax or plt.gca()
    ax.add_patch(rect)
    return rect

pdb_file = "6Y1A.pdb"
chain_id1 = "A"  # Choose the first chain for analysis
chain_id2 = "C"  # Choose the second chain for analysis

parser = PDBParser()
structure = parser.get_structure('pdb', pdb_file)
chain1 = None
chain2 = None
for c in structure.get_chains():
    if c.id == chain_id1:
        chain1 = c
    if c.id == chain_id2:
        chain2 = c

hb_plot_1 = hydrogen_bond_plot(pdb_file, chain_id1, chain_id2)


plt.imshow(hb_plot_1, cmap='binary')
residues1 = [amino_acid[residue.get_resname()] if residue.get_resname() != 'NH2' else 'NH2'  for residue in chain1 ]
residues2 = [amino_acid[residue.get_resname()] if residue.get_resname() != 'NH2' else 'NH2' for residue in chain2 ]
plt.xticks(ticks=list(range(len(residues1))), labels=residues1 )
plt.yticks(ticks=list(range(len(residues2))), labels=residues2)
for i in range(hb_plot_1.shape[0]):
    for j in range(hb_plot_1.shape[1]):
        highlight_cell(i, j, edgecolor='black', lw=1.5)
plt.show()

chain_id1 = "C"  # Choose the first chain for analysis
chain_id2 = "E"  # Choose the second chain for analysis
hb_plot_2 = hydrogen_bond_plot(pdb_file, chain_id1, chain_id2)
plt.imshow(hb_plot_1, cmap='binary')
residues1 = [amino_acid[residue.get_resname()] if residue.get_resname() != 'NH2' else 'NH2'  for residue in chain1 ]
residues2 = [amino_acid[residue.get_resname()] if residue.get_resname() != 'NH2' else 'NH2' for residue in chain2 ]
plt.xticks(ticks=list(range(len(residues1))), labels=residues1 )
plt.yticks(ticks=list(range(len(residues2))), labels=residues2)
for i in range(hb_plot_1.shape[0]):
    for j in range(hb_plot_1.shape[1]):
        highlight_cell(i, j, edgecolor='black', lw=1.5)
plt.show()


plt.imshow(hb_plot_1 + hb_plot_2)
residues1 = [amino_acid[residue.get_resname()] if residue.get_resname() != 'NH2' else 'NH2'  for residue in chain1 ]
residues2 = [amino_acid[residue.get_resname()] if residue.get_resname() != 'NH2' else 'NH2' for residue in chain2 ]
plt.xticks(ticks=list(range(len(residues1))), labels=residues1 )
plt.yticks(ticks=list(range(len(residues2))), labels=residues2)
for i in range(hb_plot_1.shape[0]):
    for j in range(hb_plot_1.shape[1]):
        highlight_cell(i, j, edgecolor='black', lw=1.5)
```

```python
from Bio import SeqIO
from Bio.SeqRecord import SeqRecord
from Bio.Seq import Seq
from Bio import pairwise2
import networkx as nx
import matplotlib.pyplot as plt

def create_similarity_network(fasta_file, threshold):
    # Parse FASTA file and extract sequences
    sequences = []
    for record in SeqIO.parse(fasta_file, "fasta"):
        sequences.append(record)

    # Create an empty graph
    G = nx.Graph()

    # Compute sequence similarities and add edges to the graph
    for i in range(len(sequences)):
        for j in range(i+1, len(sequences)):
            seq1 = sequences[i].seq
            seq2 = sequences[j].seq
            alignment = pairwise2.align.globalxx(seq1, seq2, one_alignment_only=True)[0]
            similarity = alignment[2] / max(len(seq1), len(seq2))  # Calculate sequence similarity
            if similarity >= threshold:
                G.add_edge(sequences[i].id, sequences[j].id, weight=similarity)

    return G

def plot_similarity_network(G):
    # Draw the network
    pos = nx.spring_layout(G)
    weights = [edata['weight'] for f, t, edata in G.edges(data=True)]
    nx.draw(G, pos, with_labels=True, node_size=500, node_color='skyblue', font_size=10, font_weight='bold', edge_color=weights, width=2.0, edge_cmap=plt.cm.Blues)
    plt.title("Sequence Similarity Network")
    plt.show()

# Example usage
fasta_file = "output.fasta"
threshold = 0.5  # Adjust the threshold as needed
G = create_similarity_network(fasta_file, threshold)
plot_similarity_network(G)
```

```python
from Bio.PDB import PDBParser
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord
from Bio import SeqIO
from glob import glob

def pdb_to_fasta(pdb_files, output_file):
    sequences = []

    for pdb_file in pdb_files:
        parser = PDBParser(QUIET=True)
        structure = parser.get_structure("pdb", pdb_file)

        for model in structure:
            for chain in model:
                chain_id = chain.id
                seq = ""
                for residue in chain:
                    resname = residue.resname
                    if resname == "GLY":
                        aa = "G"
                    else:
                        aa = residue.resname[:3]
                    seq += aa
                seq_record = SeqRecord(Seq(seq), id=chain_id, description=pdb_file)
                sequences.append(seq_record)

    SeqIO.write(sequences, output_file, "fasta")

pdb_files =  glob("/data/yang_lab/nehilpkd/fibrilverse_rcsb/*.pdb")
output_file = "output.fasta"  # Output FASTA file
pdb_to_fasta(pdb_files, output_file)
```
