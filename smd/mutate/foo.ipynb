```python
import enzy_htp
from enzy_htp.structure import PDBParser
from enzy_htp.mutation import assign_mutant, mutate_stru, Mutation, get_mutant_name_tag
from enzy_htp.mutation.api import sync_mutation_over_chains
import random
from copy import deepcopy

amino_acid = {'CYS': 'C', 'ASP': 'D', 'SER': 'S', 'GLN': 'Q', 'LYS': 'K',
     'ILE': 'I', 'PRO': 'P', 'THR': 'T', 'PHE': 'F', 'ASN': 'N', 
     'GLY': 'G', 'HIS': 'H', 'LEU': 'L', 'ARG': 'R', 'TRP': 'W', 
     'ALA': 'A', 'VAL':'V', 'GLU': 'E', 'TYR': 'Y', 'MET': 'M'}
```

```python
n_mutants = 5
pdb_file = "6GK3.pdb"
parser = PDBParser()


for m in range(n_mutants):
    structure = parser.get_structure(pdb_file)
    chain_id = 'A'
    # Select a random residue
    residue_id = random.choice(structure.get_chain(chain_id).residue_idxs)
    # Init the mutation class
    source_res = structure.get_chain(chain_id).find_residue_idx(residue_id).name
    mutant_res = random.choice(list(amino_acid.keys()))
    # Create list of mutations for each chain
    mutations = [Mutation(source_res, mutant_res, chain_id, residue_id) for chain_id in structure.chain_names]
    # Apply mutation
    mutated_structure = deepcopy(mutate_stru(structure, deepcopy(mutations)))
    # Save mutated structure
    parser.save_structure(
        stru=mutated_structure, 
        outfile=f"{pdb_file.split('.')[0]}_mutant_{m}.pdb"
    )
    print(''.join([str(0) if i == j else str(1) for i,j in zip(structure.sequence['A'], mutated_structure.sequence['A']) ]))
    print(f"Mutated structure saved as {pdb_file.split('.')[0]}_mutant_{m}.pdb")
```
