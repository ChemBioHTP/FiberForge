```python
pdbs = ['supp_M114.578344_jbc.M114.578344-1']

statepoint = {
    "fiberverse_directory": "/data/yang_lab/nehilpkd/fibrilverse_rcsb", 
    "type": "single", 
    "pull_chains": [0, 1], 
    "pull_constant": 1000, 
    "pull_steps": 250000, 
    "pull_rate": 0.01,
    "forcefield": "charmm27",
}
replicates = 5
statepoints = []
for pdb in pdbs:
    statepoint["pdbID"] = pdb
    statepoints.append(statepoint.copy())
if __name__ == "__main__":
    if not project.doc.get("response_table"):
        project.doc["response_table"] = {}

    for sp in statepoints:
        if "pull_rate" not in sp:
            sp["pull_rate"] = 0.01
        if "pull_constant" not in sp:
            sp["pull_constant"] = 1000
        if "pull_steps" not in sp:
            sp["pull_steps"] = 250000
        if sp['pdbID'] == '7YPG':
            sp['protofibril_distance_threshold'] = 8.0
        else:
            sp['protofibril_distance_threshold'] = 20.0
    for sp in statepoints:
        for r in range(replicates):
            sp['replicate'] = r
            if not project.find_jobs(sp):
                print("Initializing job", sp)
                project.open_job(sp).init()
                print(project.find_jobs(sp))
```
